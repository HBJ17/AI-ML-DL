```python
import pandas as pd
```

```python
#series
data = [100,103,104]
series = pd.Series(data,index=[1,2,3])
print(series)

print(series.loc[1]) #locating using by index label

print(series.iloc[1]) #locating using by index position

print(series.values) #print values

print(series[series > 100]) #filtering

print(series.mean()) #mean



#new dictionaries

calories = {"Day 1":420,"Day 2":380,"Day 3":390}
s = pd.Series(calories)
print(s)

series.loc["Day 3"] += 500

print(series.loc["Day 1"])








```
